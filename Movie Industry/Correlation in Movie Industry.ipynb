```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Configuration of the plots
```

```python
# Read the data
df = pd.read_csv(r'C:\Users\PREDATOR\Desktop\Work\Github\Data-Analyst-Portfolio\Movie Industry\movies.csv')

# First 5 rows of the data
#pd.set_option('display.max_rows', None) #un-comment to see all data within the notebook
df.head(5)
```

```python
df.describe()
```

```python
#Check data type of each column
df.dtypes
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

Missing values is not large

```python
# Fill missing numeric values with the average value of each column
for col in df.columns:
    if df[col].dtypes != 'object':
        mean_value = np.mean(df[col])
        df[col].fillna(value=mean_value, inplace=True)
```

```python
# drop any row that have missing non-numeric value
df = df.dropna()
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# The year column has some incorrect value
# Extract year in the released column for comparision
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head(5)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# drop duplicates
df.drop_duplicates().head(5)
```

Assumption:
- Budget correlation
- Company correlation

```python
# Scatter plot with budget vs. gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Regression plot for budget and gross

reg = sns.regplot(x='budget', y='gross',data =df, scatter_kws={'color':'red'}, 
            line_kws={'color':'blue'})
reg.set(xlabel='Budget for Film', ylabel='Gross Earnings')
```

budget and gross are positively correlated

```python
df.corr()
```

```python
# correlation matrix of each field to each other
correlation_matrix = df.corr(method ='pearson')
plt.title('Correlation Matrix for Numeric Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show
```

```python
# label encode every non-numeric-field

df_label_encode = df.copy()

for col in df_label_encode.columns:
    if df_label_encode[col].dtype == 'object':
        df_label_encode[col]= df_label_encode[col].astype('category')
        df_label_encode[col] = df_label_encode[col].cat.codes
        
df_label_encode.head(5)
```

```python
correlation_matrix = df_label_encode.corr(method ='pearson')
plt.title('Correlation Matrix for All Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show
```

```python
correlation=df_label_encode.corr()
correlation_pairs =correlation.unstack()
correlation_sorted = correlation_pairs.sort_values(ascending=False)
correlation_sorted
```

```python
high_corr = correlation_sorted[0.5 < (correlation_sorted)]
high_corr
```
